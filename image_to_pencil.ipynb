```python
img=cv2.imread("dog (1).webp")
img
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(True)
plt.show()
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

#inverting the gray scale img to black and white img . Inverting the pixel values means that the black pixels become white and the white pixels become black

```python
inverted_img=255-gray_img
inverted_img = cv2.bitwise_not(gray_img)
plt.imshow(inverted_img)
```

```python
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blurred)
```

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_img,inverted_img,scale=265.0)
plt.imshow(pencil_sketch)
```

```python
cv2.imwrite("dog (1).webp",pencil_sketch)
```

```python
sketch=cv2.cvtColor(pencil_sketch,cv2.COLOR_BGR2RGB)
plt.imshow(sketch)
plt.show()
```
